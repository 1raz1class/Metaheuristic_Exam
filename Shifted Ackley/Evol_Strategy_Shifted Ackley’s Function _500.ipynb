```python
# Objective function with default parameters as mentioned above
def Obj_func(x_array, ackley_list = shift_Ackley , dimension = D, bias = f_bias):
    Sum1 = 0
    Sum2 = 0
    for i in range(0,dimension) :
        z  = x_array[i] - ackley_list[i]
        Sum1 = z**2
        Sum2 = mt.cos(2*mt.pi*z)
        Sum1 += Sum1
        Sum2 += Sum2
    return -20*mt.exp(-0.2*((Sum1/dimension)**.5))  - mt.exp(Sum2/dimension)  + 20  + mt.e + bias
```
